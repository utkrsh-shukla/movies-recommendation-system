### FORMATING THE COLUMNS

```python
# genres column
movies.iloc[0]['keywords']
```

```python
def convert(obj):
    import ast
    obj = ast.literal_eval(obj)
    L = []
    for i in obj:
        L.append(i['name'])
    return L
```

```python
movies['genres'] =  movies['genres'].apply(convert)
```

```python
# keywords column
movies['keywords'] = movies['keywords'].apply(convert)
movies['keywords']
```

```python
# cast column
movies['cast'].iloc[0] 
def convert_limit_three(obj):
    import ast
    obj = ast.literal_eval(obj)
    L = []
    for i in obj[:3]:
        L.append(i['name'])
    return L 
movies['cast'] = movies['cast'].apply(convert_limit_three)
```

```python
# crew column
movies['crew'].iloc[0]
```

```python
def only_director(obj):
    import ast
    obj = ast.literal_eval(obj)
    L = []
    for i in obj:
        if( i['job'] == 'Director'):
            L.append(i['name'])
            return L 
movies['crew'] = movies['crew'].apply(only_director)
```

```python
movies.head()
```

```python
# overview column
movies['overview'] = movies['overview'].apply( lambda x : x.split(' ') )
```

### SPACE REMOVAL FOR ACCURACY OF TAGS

```python
def remove_space(obj):
    if obj == None:
        return []
    for i in range(len(obj)):
        obj[i] = obj[i].replace(' ', '')  
    return obj
```

```python
movies.head()
```

```python
movies['crew'][movies['crew'].isnull()]
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)

```

```python
movies.info()
```

### TAGS

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies['tags'] = movies['tags'].apply( lambda x : " ".join(x).lower() )
```

```python
movies['tags'][0]
```

### TEXT VECTORIZATION BY METHOD : **BAG OF WORDS**
Under this method we join all the tags then out of this huge resultant text we will extract the top 5000 most frequent non stopping words. Then every tag will be converted to a 5000 column vectors, under each column the value will show the count of the column value present in each tag. So as our data set is consist of 5000 movies we would have 5000 vectors of 5000 dimensions.
Then we will recommend movies on basis of how close the vectors are 

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
vector = cv.fit_transform(new_movie['tags']).toarray()
cv.get_feature_names_out()
```

CORRECTION : No of movies are not 5000 it was meant to be approx 5000 

```python
# Some features are numeric, as years can still be considered as a feature but '000', '007' are not useful features
# Similarly some words are repeated by plurar and singular form, so we need to handle that
# for this we will use stemming under which we will convert all words to their root form like 'running', 'ran' to 'run', 'run' respectively
# for this we will use nltk library
%pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(obj):
    L = []
    for i in obj.split():
        L.append(ps.stem(i))
    return " ".join(L)
```

```python
stem('love loved loving banker banking')
```

```python
new_movie['tags'] = new_movie['tags'].apply(stem)
```

```python
vector = cv.fit_transform(new_movie['tags'].apply(stem)).toarray()
list(cv.get_feature_names_out())
```

```python
new_movie
```

Now we have our 5000 dimensional 4800 around vectors. We need to find distance between the movies. We will go for cosine distance instead of tip to tip distance. Euclidean distance is not a good measure for high dimensionality distance as cosine make more sense. like for example if two movies are 90 degree distant that makes movies very deferrent and if its 180 its completely opposite. So 

S ∝ 1/d

S = k/d

where:
S = Similarity
d = Distance
k = Constant of proportionality

### DISTANCE CALCULATION FOR MOVIES VECTORS FROM EACH OTHER

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
similarity
```

### Now comes the main recommend function
It will take a movie in input and recommend the five similar movie wrt it

so plan is to sort the similarities in reverse order but we will loose the index so we will use enumerate function which makes every element of array a tuple by also including the index of it, so that after sort we don't loose it

```python
similarity[0]

```

```python
temp = list(enumerate(similarity[0]))
```

```python
sorted(temp, key = lambda x : x[1], reverse = True)
```

```python
def recommend(movie_title):
    index = new_movie[new_movie['title'] == movie_title].index[0]
    distance = sorted(list(enumerate(similarity[index])), key = lambda x : x[1], reverse = True)
    for i in distance[1:6]:
        print(new_movie.iloc[i[0]].title)

recommend('Harry Potter and the Half-Blood Prince')          
```
